```python
cluster_list_pattern={}
K_Mean_Error_List={}
K_Mean_Error_List_Print={}
K_Mean_Inertia_List={}



def process_clustering(pl,k_cluster, f):
    
    from sklearn.cluster import KMeans

    # Number of clusters
    kmeans = KMeans( init='random', n_clusters=k_cluster)
    # Fitting the input data
    kmeans = kmeans.fit(pl)
    # Getting the cluster labels
    labels = kmeans.predict(pl)
    # Centroid values
    centroids = kmeans.cluster_centers_
    
    
    #np.set_printoptions(formatter={"float_kind": lambda x: "%:.f2" % x})
    np.set_printoptions(precision=2)
    
    d_o =f"\n Centroids ------> {centroids} \n \n \n labels ----> {labels} {kmeans.labels_}, inertia = {kmeans.inertia_}\n"
    

    print(d_o)
    f.write(d_o)
    
    # Determine Assignment Matrix
    
    ROWS = len(pl)
    COLS = len(centroids)
    AM = np.zeros((ROWS,COLS))
    
    
    for k,cl in enumerate(labels):
        AM[k,cl] = 1  
            
            
    # Determine C R Matrix  
    CRM = centroids
        
    # Calcualte Multiplication                                                    
    CRM_DataFrame = pd.DataFrame(data=CRM)
    
    
    d_o = "\n CRM Matrix for Data Points \n"
    f.write(d_o)
    print(d_o)
    
    pd.options.display.float_format = '{:.2f}'.format
    d_o=f"{CRM_DataFrame.to_string(index=False,  show_dimensions=True)}"
    print(d_o)
    f.write(d_o)
    
    AM_DataFrame = pd.DataFrame(data=AM)
    

    d_o = f"\n\n AM Matrix for Data Points \n"
    print(d_o)
    f.write(d_o)
    
 
    d_o =f"{AM_DataFrame.to_string(header=None,index=False)}"
    print(d_o)
    f.write(d_o)
    
    PM = np.dot(AM,CRM)   
    PM_DataFrame = pd.DataFrame(data=PM)
    
    d_o="\n \n PM Matrix for Data Points \n"
    print(d_o)
    f.write(d_o)
    
    #print("PM Matrix for Data Points \n")
    header_list = [ 'X'+ str(i-0) for i in range(len(pl.columns))]
    
    d_o = f"{PM_DataFrame.to_string(header=header_list,index=False, show_dimensions=True)}"
    print(d_o)
    f.write(d_o)
    
    # Calculate Error                              
   
    K_Mean_Error = 0 
    for i in range(PATTERNS):
        for j in range(FEATURES):
           K_Mean_Error =K_Mean_Error + np.power(( pl.iloc[i,j] - PM_DataFrame.iloc[i,j]),2)

    return (K_Mean_Error, kmeans.inertia_)


K_Mean_Error_List={}

DM = pd.read_csv("engineering-mark-sheet - Sheet1.csv") 

for k in  [ 2, 5, 10, 50, 100,1000]:
#for k in  range(2,100) :
        DM = pd.read_csv("engineering-mark-sheet - Sheet1.csv")
        #DM=DM.sample(frac=1)
        #DM = DM.reset_index(drop=True)
        DM.drop(['gender','race/ethnicity','parental level of education', 'lunch','test preparation course'], axis=1, inplace=True)
        
        
        #DM.drop(DM.loc[1:990].index, inplace=True)

        
        #DM.reset_index(drop=True)
        DM.head()
        DM.describe()
        
        PATTERNS = len(DM.index)
       
        
        FEATURES = len(DM.columns)
        
        DM.head(20)
        
 
        file_name = 'file_K_MEAN_RANDOM'  + str(k) + '_.txt'
    
        f = open(file_name, 'w') 
    
  
        (K_Mean_Error, inertia_)  = process_clustering(DM, k, f)
    
        K_Mean_Error = "{:.2f}".format(K_Mean_Error)

        d_o=f"\n \n Calculated Error ==>  {K_Mean_Error } for K Mean {k}"
        
        print(d_o)

        f.write(d_o) 
        
        inertia_ = "{:.2f}".format(inertia_)
    
        K_Mean_Error_List[ k ] = K_Mean_Error
        K_Mean_Inertia_List[k] =   inertia_
        
        d_o=f"\n \n Calculated Inertia  ==>  {inertia_} for K Mean {k}"
                     
        print(d_o)

        f.write(d_o)      
              
        K_Mean_Error_List_Print[k] =  K_Mean_Error_List[ k ] 
        
f.close()

file_name = 'file_K_MEAN_RANDOM_RESULT'  + str(k) + '_.txt'
    
    
with open(file_name, 'w') as f :
    d_o = f"\n--------------- K Mean  Table - Error(MSE) ------------->"
    print(d_o)
    f.write(d_o)

    for e, v in K_Mean_Error_List_Print.items():
        d_o = f'\n   K-Mean({e}),  ======>  {v} \n'
        print(d_o)
        f.write( d_o )
f.close()

#dp = pd.DataFrame.from_dict(K_Mean_Error_List_Print)

#dp.style.set_properties(**{'text-align': 'right'})

#dp.head()



        

        
        
```
